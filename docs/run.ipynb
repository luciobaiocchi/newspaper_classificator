```python
import numpy as np
from sklearn.preprocessing import FunctionTransformer, MinMaxScaler 
from sklearn.pipeline import make_pipeline

pipe1 = make_pipeline(
    MinMaxScaler()
    )

pipe2 = make_pipeline(
    MinMaxScaler(),
    FunctionTransformer(lambda x: 1 / (x**2+1))
)
X = np.array([
    [1,2],
    [3,5],
    [2,10],
])
print (pipe1.fit_transform(X))
print (pipe2.fit_transform(X))
```
